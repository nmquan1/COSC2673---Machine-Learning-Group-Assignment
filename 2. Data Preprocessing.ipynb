```python
import os
import shutil
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
import matplotlib.pyplot as plt
import random
from pathlib import Path
from pyTools.preprocess import *
```

```python
# Define the root directory containing the leaf folders
root_dir = '../ML_Asm2_ProjFiles/trafficsigns_dataset/trafficsigns_dataset'
# Define the directory to save the copied images
preview_dir = 'preview'
preview_dir2 = 'preview2'

if os.path.exists(preview_dir):
    shutil.rmtree(preview_dir)
shutil.copytree(root_dir, preview_dir)

if os.path.exists(preview_dir2):
    shutil.rmtree(preview_dir2)
shutil.copytree(root_dir, preview_dir2)

augmented_data_shape = Path("../ML_Asm2_ProjFiles/preview")
augmented_data_char = Path("../ML_Asm2_ProjFiles/preview2")
```

```python
datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)
```

```python
parent_dir = Path(augmented_data_shape)

categories = ['Diamond', 'Round', 'Hex', 'Square', 'Triangle']
counts = [len(list(parent_dir.glob('diamond/*.png'))),
          len(list(parent_dir.glob('round/*.png'))),
          len(list(parent_dir.glob('hex/*.png'))), 
          len(list(parent_dir.glob('square/*.png'))), 
          len(list(parent_dir.glob('triangle/*.png')))]

plt.figure(figsize=(10, 6))
plt.bar(categories, counts, color='skyblue')
plt.xlabel('Categories')
plt.ylabel('Number of Images')
plt.title('Number of PNG Images in Each Category')
plt.show()
```
