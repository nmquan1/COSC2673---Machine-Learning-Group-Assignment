```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.utils import image_dataset_from_directory

# Convert the class names to a NumPy array for easy indexing
class_names = np.array(data.class_names)

# Load the first batch from the training dataset
plt.figure(figsize=(10, 10))  # Define the figure size for a grid of images
for images, labels in data.take(1):
    # Plot 9 images in a 3x3 grid
    for i in range(9):
        ax = plt.subplot(3, 3, i + 1)  # Create a subplot in a 3x3 grid
        print("Tensor shape:", images.shape)
        img = images[i].numpy().astype("uint8")  # Convert tensor to NumPy array

        # Use the numerical label to get the class name
        class_label = labels[i].numpy()  # Get the numerical label
        class_name = class_names[class_label]  # Get the corresponding class name

        plt.imshow(img, cmap='gray')  # This forces grayscale plotting

        plt.title(f"Label: {class_name}")  # Set the title with the class name
        plt.axis("off")  # Hide the axis for clarity

    plt.tight_layout()  # Adjust the layout for better spacing
    plt.show()  # Display all 9 images in a single plot
```

```python
train_size = int(len(data)*.7)
val_size = int(len(data)*.2)
test_size = int(len(data)*.1)
```

```python
train = data.take(train_size)
val = data.skip(train_size).take(val_size)
test = data.skip(train_size+val_size).take(test_size)
```

```python
plt.figure(figsize=(10, 10))
samples = 5
for images, labels in train.take(1):
    for i in range(samples):
        ax = plt.subplot(1, samples, i + 1)  # Create a subplot
        plt.imshow(images[i].numpy().astype("uint8"), cmap='gray')  # Convert tensor to numpy arr
        plt.title(class_names[labels[i]])  # Class name according to data.class_name
        plt.axis("off")  # Hide axis

plt.show()
```

***compile()***:
  - ***optimizer***: Stochastic Gradient Descent, **Adoptive Moment Estimator(Adam)**
  - ***loss***: **Categorical (Focal) Cross Entropy**
  - ***metrics***: **(Categorical) Accuracy**, Top K Categorical, Categorical Cross Entropy, F1/Precision/Recall, AUC, IoU, MeanIoU

Initial design for baseline model.
Our goal was to design a simple Convolutional Neural Network that utilizes keras libraries and utilities. Partially because of keras's versatility as a highly customisable tool, this model poses as a milestone for us to implement different modifications and tunings in development in the next steps.

```python
MLP = Sequential([
    Rescaling(1/255, input_shape=(128, 128, 1)),
    Flatten(),
    BatchNormalization(),  # Batch normalization
    Dense(256, activation='relu'),
    Dropout(0.5),
    BatchNormalization(),  # Batch normalization for stability
    Dense(256, activation='relu'),
    Dense(16, activation='softmax'),
])

MLP.compile(
    optimizer=Adam(),  #Pending: SGD and Adam no param
    loss=tf.keras.losses.SparseCategoricalCrossentropy(),
    # Sparse Categorical Cross Entropy is used when the labels are one hot encoded
    # LABELS ARE 1HOT ENCODED (wrote this so that I don't waste another hour trying to figure out why binary classification was the only thing that worked)
    metrics=['accuracy'],
)

MLP.fit(
    train,
    epochs=5,
    batch_size = 16,
    validation_data=val,
)
```

```python
# Evaluate the model
loss, accuracy = MLP.evaluate(test)
print("Test loss:", loss)
print("Test accuracy:", accuracy)
```

```python
logdir='logs'
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=logdir)
hist = MLP.fit(train, epochs=50, validation_data=val, callbacks=[tensorboard_callback])
```
