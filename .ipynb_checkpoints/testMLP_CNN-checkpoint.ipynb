```python
train_size = int(len(data)*.7)
val_size = int(len(data)*.2)
test_size = int(len(data)*.1)
```

```python
train = data.take(train_size)
val = data.skip(train_size).take(val_size)
test = data.skip(train_size+val_size).take(test_size)
```

***compile()***:
  - ***optimizer***: Stochastic Gradient Descent, **Adoptive Moment Estimator(Adam)**
  - ***loss***: **Categorical (Focal) Cross Entropy**
  - ***metrics***: **(Categorical) Accuracy**, Top K Categorical, Categorical Cross Entropy, F1/Precision/Recall, AUC, IoU, MeanIoU

Initial design for baseline model.
Our goal was to design a simple Multilayer Perceptron baseline that utilizes keras libraries and utilities. Partially because of keras's versatility as a highly customisable tool, this model poses as a milestone for us to implement different modifications and tunings in development in the next steps.

```python
#baseline model: MLP
MLP = Sequential([
    Rescaling(1/255, input_shape=(256, 256, 1)),
    Flatten(),
    Dense(256, activation='relu'),
    Dense(16, activation='sigmoid'),
])

MLP.compile(
    loss=tf.keras.losses.SparseCategoricalCrossentropy(),
    # Sparse Categorical Cross Entropy is used when the labels are one hot encoded
    # LABELS ARE 1HOT ENCODED (wrote this so that I don't waste another hour trying to figure out why binary classification was the only thing that worked)
    metrics=['accuracy'],
)

logdir='logs'
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=logdir)
hist = MLP.fit(train, 
               epochs=40, 
               validation_data=val, 
               callbacks=[tensorboard_callback])
```

```python
# Evaluate the model
loss, accuracy = MLP.evaluate(test)
print("Test loss:", loss)
print("Test accuracy:", accuracy)
```

```python
# Load your MLP model
# If you have a saved model, uncomment the following line
# MLP = load_model("your_model_path")


import cv2
import random
# Corrected function to preprocess a single image
# Function to preprocess a single image
# Corrected function to preprocess a single image
def preprocess_image(image_path):
    # Load the image
    img = Image.open(image_path)
    # Convert to grayscale
    img = img.convert("L")
    # Resize to 28x28
    img = img.resize((256, 256))
    # Normalize
    img = np.array(img) / 255  # Scale pixel values to 0-1
    # Reshape to (28, 28, 1) for the model
    img = img.reshape((256, 256, 1))
    return img

# Get raw data
data_dir = "../ML_Asm2_ProjFiles/tsrd-train"  # Path to your dataset
raw_data = []

# Collect and preprocess images
for img_file in os.listdir(data_dir):
    if img_file.endswith((".png", ".jpg", ".jpeg", ".bmp")):
        img_path = os.path.join(data_dir, img_file)
        preprocessed_img = preprocess_image(img_path)
        raw_data.append(preprocessed_img)

# Shuffle the images in raw_data
random.shuffle(raw_data)

# Convert the list to a NumPy array
raw_data = np.array(raw_data)

# Check the number of images and the shape of the first image
print("Number of images:", len(raw_data))  # Check the number of images
print("Shape of first image:", raw_data[0].shape)  # Ensure it's (28, 28, 1)

# Get predictions from the model
predictions = MLP.predict(raw_data)

# Interpret the predictions
predicted_classes = np.argmax(predictions, axis=1)  # Get predicted classes

print("Predicted Classes:", predicted_classes)  # Print the predicted class for each image
```
